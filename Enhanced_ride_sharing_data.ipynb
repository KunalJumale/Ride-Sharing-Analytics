```python
# Q. What’s the average fare per payment method for completed rides only?

df[df['CancellationStatus'] == 'Completed'].groupby('PaymentMethod')['TotalFare'].mean()
```

```python
# Q. Sort the dataframe so that: First: Completed rides only Then: Highest rated drivers first
df[df["CancellationStatus"]=="Completed"].sort_values("DriverRating", ascending=False)
```

```python
# Q. Average fare by VehicleType
df.groupby("VehicleType")["TotalFare"].mean()
```

```python
# Q. Sort rides by TotalFare in descending order
df.sort_values("TotalFare", ascending=False)
```

```python
# Q. Find all rides where: City is Delhi Payment Method is Cash DriverRating is 4 or higher
df[(df["City"]=="Delhi") & (df["PaymentMethod"]=="Cash") & (df["DriverRating"]>=4)]
```

```python
# Q. Count of rides grouped by VehicleType and CancellationStatus
df.groupby("VehicleType")["CancellationStatus"].count()
sns.countplot(x="CancellationStatus", data=df, hue="VehicleType")
plt.show()
```

```python
# 🎯 Q. List rides where DriverRating is less than 3
df[(df["DriverRating"]<3)]
```

```python
# Q14. Maximum, minimum, and average trip duration
df["TripDurationMin"].mean()
```

```python
# Q. Sort by City and then by TripDistanceKM (within each city)
df.sort_values("TripDistanceKM", ascending=False)
```

```python
# Q. Count number of rides per city
df["City"].value_counts()
sns.countplot(x="City", data=df)
plt.show()
```

```python
#. Count of payment methods used
df["PaymentMethod"].value_counts()
sns.countplot(x="PaymentMethod", data=df)
plt.show()
```

```python
# Q. Create a column FarePerMinute = TotalFare ÷ TripDurationMin
df["FarePerMinute"]=df["TotalFare"]/df["TripDurationMin"]
df
```

```python
# Q. Find the average, min, and max TotalFare by City
df.groupby("City")["TotalFare"].agg(["min", "mean", "max"])
```

```python
# Q. Sort by highest FarePerMinute (to find expensive short rides)
df.sort_values("FarePerMinute", ascending=False).head()
```

```python
# 🎯 Q. Filter all rides with TripDistanceKM between 10 and 20
df[(df["TripDistanceKM"]>10) & (df["TripDistanceKM"]>20)]
```

```python
# Q. Filter all rides in Mumbai using Sedan
df[(df["City"]=="Mumbai") & (df["VehicleType"]=="Sedan")]
```

```python
# Q. Show the ride with the highest TotalFare
df["TotalFare"].max()
```

```python
# Q. Add a new column TripSpeedKMH = Distance ÷ (Duration in hours)
df["TripSpeedKMH"]=df["TripDistanceKM"]/df["TripDurationMin"]
df
```
